# MODELO SIR(Suscetível infectado recuperado)

$$\frac{dS}{dt} = -β\frac{SI}{N}$$

$$\frac{dI}{dt} = β\frac{SI}{N}-γI$$

$$\frac{dR}{dt} = γI$$


```python
import numpy as np
import matplotlib.pyplot as plt


deltat = 0.01
w = 3

def velocidade(t, x, v):
  return v

def aceleracao(t, x, v):
  return -w*x

t = np.zeros(10000)
v = np.copy(t)
x = np.copy(t)

v[0] = 1
x[0] = 2
t[0] = 0


for i in range(len(t)-1):
  v[i+1] = v[i] + deltat*aceleracao(t[i], x[i], v[i])
  x[i+1] = x[i] + deltat*v[i]
  t[i+1] = t[i] + deltat


import numpy as np
import matplotlib.pyplot as plt


deltat = 0.01
w = 3


def aceleracao(t, x, v):
  return -w*x

t_cromer = np.zeros(10000)
v_cromer = np.copy(t)
x_cromer = np.copy(t)

v_cromer[0] = 1
x_cromer[0] = 2
t_cromer[0] = 0


for i in range(len(t)-1):
  v_cromer[i+1] = v_cromer[i] + deltat*aceleracao(t_cromer[i], x_cromer[i], v_cromer[i])
  x_cromer[i+1] = x_cromer[i] + deltat*v_cromer[i+1]
  t_cromer[i+1] = t_cromer[i] + deltat
```

```python
from matplotlib.collections import LineCollection
points = np.array([x_cromer, v_cromer]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

fig, ax = plt.subplots()

# Create a continuous norm to map from data points to colors
norm = plt.Normalize(t.min(), t.max())
lc = LineCollection(segments, cmap='viridis', norm=norm)
# Set the values used for colormapping
lc.set_array(t)
lc.set_linewidth(4)
line = ax.add_collection(lc)
fig.colorbar(line, ax=ax)

ax.set_xlim(-3, 3)
ax.set_ylim(-4, 4)
fig.patch.set_facecolor('white')
plt.gcf().set_size_inches(8, 5)

ax.set_ylabel(r'$v(t)$', fontsize=14)
ax.set_xlabel(r'$x(t)$', fontsize=14)
ax.set_title('Método de Euler-Cromer')


plt.show()
```
